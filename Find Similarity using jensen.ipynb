# Finding Similarity of two images using histogram with OpenCv in python 

### Importing libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Jensen Function for Similarity Assessing

```python
def jensen(p, q):
    # normalizing the histograms 
    newP = p/np.sum(p)
    newQ = q/np.sum(q)
    t1 = newP *np.log2(newP)
    t2 = newQ * np.log2(newQ)
    t3 = ( newP + newQ) * np.log2((newP + newQ) / 2)
    
    j = (0.5)*np.sum(t1 + t2 - t3)
  
    return j
    
```

## 1. Evaluating the similarity of an image to itself

#### Read an image using OpenCv library.

```python
pic1 = cv2.imread('C:/Users/Desktop/pic1.jpg')
plt.imshow(pic1)
```

### Calculating the histogram of the picture using OpenCv

```python
hist_pic1 = cv2.calcHist([pic1],[0], None, [256], [0,256])
hist_pic1 = hist_pic1 + 1 
```

### Calling the jensen method to assess the similarity

```python
j = jensen(hist_pic1, hist_pic1)
j
```

## Conclusion 
 As we see the j parameter for two similar pictures equals to zero.
